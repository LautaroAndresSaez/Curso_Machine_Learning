## Matriz de distancia

```python
import pandas as pd
from scipy.spatial import distance_matrix 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('./datasets/movies/movies.csv', sep=';')
data.head()
```

```python
movies = data.columns.values.tolist()[1:]
```

```python
dd_1 = distance_matrix(data[movies], data[movies], p=1)
dd_2 = distance_matrix(data[movies], data[movies], p=2)
dd_10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
def dmToDF(dm, col_name):
    return pd.DataFrame(dm, index = col_name, columns=col_name)
```

```python
dmToDF(dd_1, data['user_id'])
```

```python
dmToDF(dd_2, data['user_id'])
```

```python
dmToDF(dd_10, data['user_id'])
```

## Clustering Jerárquico Con Python

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data[movies]
```

```python
i = 'median'
Z = linkage(data[movies], i)
Z
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendrograma jerárquico para el Clustering mediante el metodo de ' + i)
plt.xlabel('ID delos usuarios')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.savefig('Dendograma ' + i )
```
