## entradas del metodo *read_csv*

- *filepath* es el nombre de la url del fichero
- *sep* indica el separador del dataset por defecto carga ','
- *dtype* permite formatear una columna a otro tipo de variable por defecto None, recibe un diccionario
- *header* por defecto "None", el argumento puede ser un numero entero, que es el numero de fila
- *names* indica el nombre de las columnas, recibe una lista 
- *skiprows* es un entero o una lista, salta un determinado numero de filas 
- *index_col* puede ser un numero o filas
- *skip_blank_lines* si es True salta los valores que tienen espacios en blanco
- *na_filter* detecta los valores que faltan, y desaparece toda la fila si se encuentra en true

## metodo open 

```python
data_3 = open(main_path+'customer-churn-model/Customer Churn Model.txt','r')

```

```python
cols = data_3.readline().strip().split(',')
n_cols = len(cols)
n_cols
```

```python
counter = 0
main_dicc = {}
for col in cols:
    main_dicc[col]=[]
```

```python
for line in data_3:
    values = line.strip().split(',')
    for i in range(n_cols):
        main_dicc[cols[i]].append(values[i])
    counter+=1
print('El data set tiene %d de filas y %d de columnas' %(counter,n_cols))
```

```python
df_3 = pd.DataFrame(main_dicc)
df_3.head()
```

### Escritura y lectura con el metodo open

```python
in_file= main_path+'customer-churn-model/Customer Churn Model.txt'
out_file = main_path+'customer-churn-model/Tab Customer Churn Model.txt'
```

```python
with open(in_file,"r") as in_file1:
    with open(out_file,"w") as out_file1:
        for line in in_file1:
            fields = line.strip().split(',')
            out_file1.write('\t'.join(fields))
            out_file1.write('\n')
```

```python
data_4 = pd.read_csv(out_file, sep="\t")
data_4.head()
```

```python
import urllib3


http = urllib3.PoolManager()
url = 'http://winterolympicsmedals.com/medals.csv'
r = http.request('GET', url)
response = r.data


#decodifico el archivo binario a string
str_data = response.decode('utf-8')

#separo las lineas
lines = str_data.split('\n')

#obtengo el nombre y numero de columnas
col_names = lines[0].split(',')
n_col = len(col_names)

#genero el diccionario de almacenamiento
counter = 0
main_diccionario = {}

for col_name in col_names:
    main_diccionario[col_name] = {}

#agrego los valores al diccionario

for line in lines:
    if(counter>0):
        values = line.strip().split(',')
        for i in range(n_col):
            main_diccionario[col_names[i]].append(values[i])
    counter+=1
```
