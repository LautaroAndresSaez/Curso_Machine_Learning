## Support Vector Machine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns
sns.set()
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.6)
```

```python
plt.scatter(X[:,0], X[:,1], c = Y, s=50, cmap='autumn')
```

```python
xx = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c = Y, s=50, cmap='autumn')
plt.plot([0.5], [2.1], 'x', color='b')

for a, b in [(1, 0.65), (0.5, 1.6), (-.2, 2.9)]:
    yy = a * xx + b
    plt.plot(xx, yy, '-k')
plt.xlim(-1, 3.5)
```

## Maximización del margen

```python
xx = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c = Y, s=50, cmap='autumn')
plt.plot([0.5], [2.1], 'x', color='b')

for a, b, d in [(1, 0.65, .33), (0.5, 1.6, .55), (-.2, 2.9, .2)]:
    yy = a * xx + b
    plt.plot(xx, yy, '-k')
    plt.fill_between(xx, yy-d, yy+d, edgecolor='none', color='#cccccc', alpha=.5)
plt.xlim(-1, 3.5)
```

## Creacion del modelo SVM

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear', C=1e10)
model.fit(X, Y)
```

```python
w = model.coef_[0]
a = - w[0] / w[1]
b = - model.intercept_[0] / w[1]
```

```python
def plt_svc(model, ax= None, plot_support=True):
    """Plot de la funcion de la decicsion para una clasificacion en 2D con SVC """
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    #Generamos los valores a evaluar
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(yy, xx)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    #representamos las fronteras y los margenes del SVC
    ax.contour(X, Y, P, colors='k', levels=[-1, 0, 1], alpha=.5,
                linestyle=['--', '-', '--'])
    
    if plot_support:
        ax.scatter(model.support_vectors_[:,0],
                   model.support_vectors_[:,1],
                  s=50,
                  linewidth=1,
                  facecolors='k')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='autumn', s=50)
plt_svc(model)
```

```python
def plt_svm(N=10, ax=None):
    X, Y =  make_blobs(n_samples=300, centers=2, random_state=0, cluster_std=.6)
    X = X[:N]
    Y = Y[:N]
    model = SVC(kernel='linear', C=1e10)
    model.fit(X,Y)
    ax = ax or plt.gca()
    ax.scatter(X[:,0], X[:,1], c=Y, cmap='autumn', s=50)
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    
    plt_svc(model, ax)
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))
fig.subplots_adjust(left=0.0625, right = 0.95, wspace = 0.1)
for ax_i, N in zip(ax, [60, 120]):
    plt_svm(N, ax_i)
    ax_i.set_title('N={0}'.format(N))
```

```python
from ipywidgets import interact, fixed
```

```python
interact( plt_svm, N=[10, 20, 50, 100, 200, 300], ax=fixed(None) )
```
