## Generacion de numeros aletorios

```python
#forma clasica entre 0 1 con decimales
np.random.random()
```

```python
def randListGenerator(dimension, a, b):
    rand_list = []
    for i in range(dimension):
        rand_list.append(np.random.random()*(b-a) + a)
    return rand_list
```

```python
randListGenerator(10,5,10)
```
