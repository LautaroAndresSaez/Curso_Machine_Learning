## SVM de Regresi√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
X = np.sort(5* np.random.rand(200,1), axis=0 )
Y = np.sin(X).ravel()
Y[::5] += 3*(0.5 - np.random.rand(40))
```

```python
plt.scatter(X, Y, color='darkorange', label='datos')
```

```python
C = 1e3
svr_lineal = SVR(kernel='linear', C=C)
svr_rbf = SVR(kernel='rbf', C=C, gamma=.1)
svr_poly = SVR(kernel='poly', C=C, degree=3)
```

```python
y_linear = svr_lineal.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_poly = svr_poly.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X, Y, color='darkorange', label='Data')
plt.plot(X, y_linear, color='navy', lw=lw, label='Lineal')
plt.plot(X, y_rbf, color='c', lw=lw, label='rbf')
plt.plot(X, y_poly, color='cornflowerblue', lw=lw, label='poly')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('SVR')
```
